```python
import numpy as np
import matplotlib.pyplot as plt
###############
# Visualization
###############
CLASSES_NAME = {0: 'Benign', 1: 'Malignant'}

rand_id = np.random.choice(range(datas_raw.shape[0]), size=10, replace=False)
datas_plt = np.array([datas_raw[i] for i in rand_id])
labels_plt = [labels[i] for i in rand_id]
labels_name = [CLASSES_NAME[name] for name in labels_plt]
plt.figure(figsize=(15, 7))
for iImgsPlt in range(10):
    plt.subplot(2, 5, iImgsPlt + 1)
    plt.imshow(datas_plt[iImgsPlt].reshape(64, 64), cmap='gray')
    plt.title('Class: %s \n Labels: %s' % (labels_name[iImgsPlt], labels_plt[iImgsPlt]), size=15)
plt.show()
```
